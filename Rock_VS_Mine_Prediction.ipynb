```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("sonar.all-data.csv",header=None)
df.head()
```

```python
df[60]=df[60].apply(lambda x:1 if x=='R' else 0)
```

```python
df.groupby(60).mean()#0 for Mine 1 for Rock
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(data=df)
plt.show()#There are many outliers in the dataset but here not going to remove them because in quality prediction somewhere it is very intense vale in prediction 
#for rock vs mine
```

```python
X=df.drop(60,axis=1)
Y=df[60]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,test_size=0.3,random_state=75)
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X_train,Y_train)
lg_pred=lg.predict(X_test)
from sklearn import metrics
print(metrics.accuracy_score(lg_pred,Y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
rf_pred=rf.predict(X_test)
print(metrics.accuracy_score(rf_pred,Y_test))
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,Y_train)
svm_pred=svm.predict(X_test)
print(metrics.accuracy_score(svm_pred,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
dt_pred=dt.predict(X_test)
print(metrics.accuracy_score(dt_pred,Y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train,Y_train)
knn_pred=knn.predict(X_test)
print(metrics.accuracy_score(knn_pred,Y_test))
```

```python
#It is clear very much from the accuracy score that Random Forest and Support vector Machine proven to be very good model for ROCK
#Vs Mine Prediction.
```
